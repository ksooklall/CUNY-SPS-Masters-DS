```python
idx = ddf.columns.get_loc('k__Archaea')
meta_df = ddf.iloc[:, 0:idx]
sample_df = ddf.iloc[:, idx:]
```

```python
meta_df.shape
```

Drop all NaN columns

```python
cols = [i for i in meta_df.columns if meta_df[i].nunique() < 2] + ['oral_anti-diabetic_medication', 'years_in_sweden']
len(cols)
meta_df = meta_df.drop(cols, axis=1).replace(['n','impaired_glucose_tolerance',"t2d"],['No Diabetes','Pre-Diabetes',"Type 2 Diabetes"])
```

```python
meta_df.shape
```

All values for species >=0, so a column with a max entry of zero means has all zeroes

```python
cols = (sample_df.max() != 0)
```

```python
sample_df= sample_df[cols[cols].index]
```

```python
sample_df.head()
```

```python
train_df = meta_df.join(sample_df)
```

```python
s = train_df['sampleID'].map(train_df.set_index('sampleID')['tcho'])
train_df['cholesterol']= train_df['cholesterol'].mask(train_df['cholesterol'].isnull(), s)
```
